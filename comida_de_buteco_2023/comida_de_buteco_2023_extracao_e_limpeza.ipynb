```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Cria listas para armazenar as informações
nomes = []
enderecos = []
links_imagens = []


for i in range(1,12):
# URL do site
  url ="https://comidadibuteco.com.br/category/butecos/rio-de-janeiro/page/{}/".format(i)

# Realiza a requisição HTTP
  response = requests.get(url)

# Cria o objeto BeautifulSoup
  soup = BeautifulSoup(response.text, "html.parser")

# localizar os cards

  cards = soup.find_all("div", class_="result-inner text-left")

# Itera sobre cada card e extrai as informações
  for card in cards:
    nome = card.find("h2").text
    endereco = card.find("p").text
    link_imagem = card.find("img")["src"]
    
    # Adiciona as informações nas listas
    nomes.append(nome)
    enderecos.append(endereco)
    links_imagens.append(link_imagem)

# Cria o dataframe com as informações
df = pd.DataFrame({
    "nome_do_bar": nomes,
    "endereco": enderecos,
    "link_da_Imagem": links_imagens
})

# Imprime o dataframe
df
```

```python

for i in range(0,len(df['endereco'])):
  sentenca = df['endereco'][i]
  sentenca = sentenca.replace("\t", "")
  sentenca = sentenca.strip()
  df['endereco'][i] = sentenca



```

```python
import re

lista = df['endereco']

valores_extraidos = []

# Utiliza expressões regulares para encontrar a parte desejada de cada string na lista
padrao = re.compile(r"\| (.*?),")
for item in lista:
    resultado = padrao.search(item)

    # Verifica se o resultado foi encontrado
    if resultado:
        # Extrai a parte encontrada
        parte_desejada = resultado.group(1)

        valores_extraidos.append(parte_desejada)

df['bairros'] = valores_extraidos
```

```python
def remover_acentos(palavra):
    tabela = {
        ord('A'): 'a', ord('a'): 'a',
        ord('É'): 'e', ord('é'): 'e',
        ord('Ê'): 'e', ord('ê'): 'e',
        ord('Í'): 'i', ord('í'): 'i',
        ord('Ó'): 'o', ord('ó'): 'o',
        ord('Ô'): 'o', ord('ô'): 'o',
        ord('Ú'): 'u', ord('ú'): 'u',
        ord('Ç'): 'c', ord('ç'): 'c',
        ord('Ñ'): 'n', ord('ñ'): 'n'
    }
    sentenca_sem_acentos = sentenca.lower().translate(tabela)
    return sentenca_sem_acentos

lista_sentencas = df['nome_do_bar']

# Percorre cada palavra da lista, remove os acentos e coloca hífen entre os espaços em branco
lista_sentenca_com_hifen = []
for sentenca in lista_sentencas:
    # Remove os acentos da palavra
    sentenca_sem_acentos = remover_acentos(sentenca)
    # Coloca hífen entre os espaços em branco da palavra
    palavra_com_hifen = sentenca_sem_acentos.replace(' ', '-')
    # Adiciona a palavra com hífen na lista de palavras
    lista_sentenca_com_hifen.append(palavra_com_hifen)

df['nome_link'] = lista_sentenca_com_hifen

df
```

```python
import pandas as pd 
df2=pd.read_csv('https://raw.githubusercontent.com/Siri2191/study_area/main/comida_de_buteco_2023/comida_de_buteco_2023',index_col = False)
df2
```

```python
import requests
from bs4 import BeautifulSoup

pratos = []
descricoes = []

def remover_acentos(palavra):
    tabela = {
        ord('Á'): 'a', ord('á'): 'a',
        ord('É'): 'e', ord('é'): 'e',
        ord('Ê'): 'e', ord('ê'): 'e',
        ord('Í'): 'i', ord('í'): 'i',
        ord('Ó'): 'o', ord('ó'): 'o',
        ord('Ô'): 'o', ord('ô'): 'o',
        ord('Ú'): 'u', ord('ú'): 'u',
        ord('Ç'): 'c', ord('ç'): 'c',
        ord('Ñ'): 'n', ord('ñ'): 'n',
        ord('Õ'): 'o', ord('ã'): 'a',
        ord('â'): 'a', ord("'"): ''

    }
    sentenca_sem_acentos = palavra.lower().translate(tabela)
    return sentenca_sem_acentos

# adicionando nomes de links sem acentos na coluna "nome_link_sem_acento"
df2['nome_link_sem_acento'] = df2['nome_link'].apply(remover_acentos)

for sufx in df2['nome_link_sem_acento']:
    url1 = f'https://comidadibuteco.com.br/{sufx}'
    url2 = f'https://comidadibuteco.com.br/{sufx}-2/'
    try:
        response = requests.get(url1)
    except:
        response = requests.get(url2)
    soup = BeautifulSoup(response.content, 'html.parser')
    nome_prato = soup.find('strong')
    descricao_prato = soup.find('br')
    if nome_prato is not None and descricao_prato is not None:
        pratos.append((nome_prato.text))
        descricoes.append((descricao_prato.text))
    else:
        pratos.append(None)
        descricoes.append(None)
        print(f"Elementos não encontrados no link: {url1} ou {url2}")

# mostrar lista de pratos
print(pratos)


```
